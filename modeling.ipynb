```python
from sklearn.model_selection import train_test_split
train_df, temp_df = train_test_split(df, test_size=0.05, random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)



print(train_df.shape, val_df.shape, test_df.shape)
```

```python
test_df.to_csv('test.csv', index=False)
val_df.to_csv('val.csv',index=False)
```

```python
from datasets import Dataset

train_dataset = Dataset.from_pandas(train_df[['input_text', 'target_text']].reset_index(drop=True))
val_dataset = Dataset.from_pandas(val_df[['input_text', 'target_text']].reset_index(drop=True))
test_dataset = Dataset.from_pandas(test_df[['input_text', 'target_text']].reset_index(drop=True))
```

```python
from transformers import T5Tokenizer
tokenizer = T5Tokenizer.from_pretrained("t5-base")

```

```python
def tokenize(row):
    input_enc = tokenizer(
        row["input_text"],
        padding="max_length",
        truncation=True,
        max_length = 512
    )

    target_enc = tokenizer(
        row["target_text"],
        padding="max_length",
        truncation=True,
        max_length = 64
    )

    labels = [
        (token if token != tokenizer.pad_token_id else -100)
        for token in target_enc["input_ids"]
    ]

    return {
        "input_ids": input_enc["input_ids"],
        "attention_mask": input_enc["attention_mask"],
        "labels": target_enc["input_ids"]
    }
```

```python
tokenized_train = train_dataset.map(tokenize, batched=False)
tokenized_val = val_dataset.map(tokenize, batched=False)
tokenized_test = test_dataset.map(tokenize, batched=False)
```

```python
from transformers import T5ForConditionalGeneration, Trainer, TrainingArguments
import torch

print("GPU available:", torch.cuda.is_available())
print("Device:", torch.device("cuda" if torch.cuda.is_available() else "cpu"))

model = T5ForConditionalGeneration.from_pretrained("t5-base") 
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

```

```python
from evaluate import load

rouge = load("rouge")

def compute_metrics(eval_preds):
    preds, labels = eval_preds.predictions, eval_preds.label_ids
    # Flatten if nested
    if isinstance(preds[0][0], list):
        preds = [p[0] for p in preds]
    # Mask and decode labels
    labels = [
        [tok if tok != -100 else tokenizer.pad_token_id for tok in lab]
        for lab in labels
    ]
    dec_preds = tokenizer.batch_decode(preds, skip_special_tokens=True)
    dec_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)
    return rouge.compute(predictions=dec_preds, references=dec_labels, use_stemmer=True)
```

```python



from transformers import Seq2SeqTrainer, Seq2SeqTrainingArguments

training_args = Seq2SeqTrainingArguments(
    output_dir="./results",
    evaluation_strategy="steps",    # or "epoch"
    # eval_steps=500,                 # run every 500 steps (not every epoch)
    save_strategy="no",
    # save_steps=500,
    predict_with_generate=False,     # ← enable generation
    generation_max_length=64,
    generation_num_beams=2,         # beam search for better quality
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    learning_rate=1e-4,
    num_train_epochs=3,
    logging_steps=500,
    load_best_model_at_end=False,
    metric_for_best_model="eval_loss",
    report_to="none",
    warmup_ratio = 0.1,
    lr_scheduler_type="cosine"
)

trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,   # see next section
    tokenizer=tokenizer,
    # compute_metrics=compute_metrics     # your fixed function
)

```

```python
import os
# import wandb
os.environ["WANDB_DISABLED"] = "true"



trainer.train()
```
